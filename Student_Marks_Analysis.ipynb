**Create a list named marks by getting 5 subject marks from the user using map() and list()**

```python
marks =list(map(int,input("enter the marks: ").split(",")))
marks
```

**• Find total marks using sum()**

```python
total_marks = sum(marks)
print(total_marks)
```

**• Find average marks using**

```python
average = total_marks / len(marks)
print(average)
```

**• Calculate over all percentage assuming each subject is out of 100**

```python
percentage = (total_marks / 500) * 100
print(percentage)
```

**Display Output**

```python
print(marks)
print(total_marks)
print(average)
print(percentage )
```
