## ** Student Marks Filter System using continue**

```python
# Print students who scored below 50 using continue

students = ["Vishnu", "Sudhan", "Mani", "Durga", "Devi"]
marks = [45, 78, 32, 66, 49]

i = 0

while i < len(students):

    # Skip students scoring 50 or above
    if marks[i] >= 50:
        i += 1
        continue

    print(f"{students[i]} scored {marks[i]} (Below 50)")

    i += 1
```
