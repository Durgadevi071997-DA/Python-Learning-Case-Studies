##  Password Verification System with Limited Attempts

```python
# Original password
password = "durga123"

# Attempt variables
attempts = 0
max_attempts = 3

# while loop
while attempts < max_attempts:

    # Taking input
    user_input = input("Enter password: ")

    # Checking password
    if user_input == password:
        print("Access granted!")
        break

    else:
        attempts += 1
        print(f"Incorrect password, try again. ({max_attempts - attempts} attempts left)")

# Maximum attempts condition
if attempts == max_attempts:
    print("Maximum attempts reached. Access denied!")
```
