```python
# String Analysis using List Comprehension

text = "data analytics"

#  uppercase
upper_text = [x.upper() for x in text]

# Extract vowels
vowels = [x for x in text if x in "aeiou"]

# Extract consonants
consonants = [x for x in text if x.isalpha() and x not in "aeiou"]

# Remove spaces
remove_spaces = [x for x in text if x != " "]

print("Uppercase Letters :", upper_text)
print("Vowels            :", vowels)
print("Consonants        :", consonants)
print("Without Spaces    :", remove_spaces)
```
