Toys Store Management

create a list named  Toy Store with initial items: Teddy Bear, Doll, Puzzle, slide,


```python
store = ["Teddy Bear", "Doll", "Puzzle", "Slide"]
print(store)
```

• Add multiple items using extend()

```python
store.extend(["Phone", "Gun","Train",])
print(store)
```

Remove item Puzzle using remove()

```python
store.remove("Puzzle")
print(store)
```

 Insert item at index 2 using insert()

```python
store.insert(2,"Bat")
print(store)
```

• Remove last item using pop()

```python
store.pop()
print(store)
```

• index() to find position of Bat

```python
print("position of Bat :" , store.index("Bat"))
```

• count() to count occurrences of Doll

```python
print("Count of Doll:", store.count("Doll"))
```

• Display first item




```python
print("First item:", store[0])
```

• Display last item

```python
print("Last item:" , store[-1])
```

Display First Three

```python
print("first three :",store[:3])
```

Find total number of items using len()

```python
print("Total of item:", len(store))
```

Display All Output

```python
store = ["Teddy Bear", "Doll", "Bat", "Slide", "Phone", "Gun"]

print("Toys store:", store)

print("Position of Bat :", store.index("Bat"))

print("Count of Doll:", store.count("Doll"))

print("First Item:", store[0])

print("Last Item:", store[-1])

print("First 3 Items:", store[:3])

print("Total of Items:", len(store))
```
