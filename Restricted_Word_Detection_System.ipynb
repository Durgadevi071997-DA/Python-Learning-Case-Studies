```python
#total number of 6 in list
numbers = [2, 4, 6, 8, 10, 6, 0 ,6 ]

count =0

for num in numbers:
    print(num)
    if num == 6:
        print("Number found! in the loop.")
        count +=1
        continue
        print()
    else:
      print("Still searching")

print("Total num of 6 in my list",count)
```
