```python
# Create lists
students = ["Anu", "Ravi", "Kiran", "Priya", "Arun"]
marks = [85, 90, 78, 92, 88]

# Using range()
print("===== Using range() =====")
for i in range(len(students)):
    print("Student:", students[i], "| Marks:", marks[i])

# Using enumerate()
print("\n===== Using enumerate() =====")
for index, name in enumerate(students):
    print("Index:", index, "| Student Name:", name)

# Using zip()
print("\n===== Using zip() =====")
for name, mark in zip(students, marks):
    print("Student:", name, "| Marks:", mark)

# Final Result
print("\n===== Final Result =====")
for name, mark in zip(students, marks):
    print(f"{name} scored {mark} marks")
```
