**Student Data Management System**

```python
student = {
    "name": ["durga", "mani", "vishnu"],
    "age": [21, 22, 23],
    "course": ["Data Science", "AI", "Machine Learning"]
}
```

**Add new one student **

```python
student["name"].append("David")
student["age"].append(24)
student["course"].append("Python")
```

**Display all Data**

```python
print("Names:", student["name"])
print("Ages:", student["age"])
print("Courses:", student["course"])
```
