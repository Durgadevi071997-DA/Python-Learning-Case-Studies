# **ATM PIN Verification System**

```python
# Correct PIN
correct_pin = "1207"

# while loop
while True:

    # Taking input from user
    user_pin = input("Enter ATM PIN: ")

    # Checking PIN
    if user_pin == correct_pin:
        print("Login Successful")
        break

    else:
        print("Incorrect PIN... Try Again")
```
