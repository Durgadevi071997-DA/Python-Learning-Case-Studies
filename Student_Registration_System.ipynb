**Student Registration System**

```python
# Studend Details
student_name = input(" enter student name :")
student_age = int(input("enter student age :"))
course_name = input("enter course name :")
course_fee = float(input("enter course fee :"))

# Calculations
gst = course_fee * 0.18
total_fee = course_fee + gst

print(f"Student Name: {student_name}")
print(f"Course Fee: {course_fee}")
print(f"GST (18%): {gst}")
print(f"Total Fee: {total_fee}")
```

```python
# Displaying Datatypes

print(type(student_name))
print(type(student_age))
print(type(course_fee))
```

**Student_ID_Card_Generation**

```python
student_name = input ("enter student name :")
roll_num = input ("enter roll num :")
department_name = input ("enter department name :")
```

```python
student_name = student_name.upper()
department_name = department_name.upper()
```

```python
student_name = student_name[0:5]
department_name = department_name[0:5]
```

```python
student_id = student_name + roll_num + department_name
```

```python
print(f"student_Name : {student_name}")
print(f"department_name : {department_name}")
print(f"roll_num : {roll_num}")
print(f"student_id : {student_id}")
```
