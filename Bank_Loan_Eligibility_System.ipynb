**Bank Loan Eligibility System**

** Customer Data**

```python
customer_name = "Durgadevi"
age = 29
income = 55000
credit_score = 720
existing_loan = False
```

**Loan Eligibility Check**


```python
if age >= 21 and income >= 30000:

    if credit_score >= 700 and not existing_loan:
        status = "Approved"
    elif credit_score >= 600 and income >= 40000:
        status = "Conditionally Approved "
    else:
        status = "Rejected "
else:
    status = "Not Eligible "
```

**Display Result**

```python
print("Customer:", customer_name)
print("Age:", age)
print("Income:", income)
print("Credit Score:", credit_score)
print("Loan Status:", status)
```
