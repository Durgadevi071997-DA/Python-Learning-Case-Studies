**# Creating set**

```python
courses = {"Economics", "English", "Computer Science", "Commerce", "Accounts"}
```

**# Display original set**

```python
print("Original Courses:", courses)
```

**# add() - Add one course**

```python
courses.add("History")
print("add():", courses)
```

**# update() - Add multiple courses**

```python
courses.update(["AI", "Machine Learning"])
print("update():", courses)
```


**# remove() - Remove course **

```python
courses.remove("Commerce")
print("remove():", courses)
```

**# discard() - Remove safely**

```python
courses.discard("Java")
print("discard():", courses)
```

**# Display total courses**

```python
print("Total Courses:", len(courses))
```
