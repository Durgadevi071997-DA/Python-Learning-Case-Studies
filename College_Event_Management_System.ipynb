# **1. Participants List**

**Create participants**

```python
participants = ["durga","devi","mani","vishnu"]
```

**Append()**

```python
participants.append("sudhan")
print(participants)
```

**Insert()**

```python
participants.insert(1,"kowshik")
print(participants)
```

**Remove()**

```python
participants.remove("durga")
print(participants)
```

**display output**

```python
print("Participants List:", participants)
```

# **2. Events Tuple List**

```python
Events = ("Dance","Drawing","singing","drama")
```

**First Events**

```python
first_Events = Events[0]
print(first_Events)
```

**Last Events**

```python
Last_Events = Events[-1]
print(Last_Events)
```

**Total Events**

```python
Total_Events =len(Events)
print(Total_Events)
```

# **3. Departments Set**

**Create list**

```python
Departments = ["Economics","Tamil","Physics","Chemistry","Id"]
```

**Convert List into set**

```python
dept_set = set(Departments)
print(dept_set)
```

**Add**

```python
dept_set.add("Maths")
print(dept_set)
```

**Remove**

```python
dept_set.discard("Id")
print(dept_set)
```

**Display Output**

```python
print("Departments set :", dept_set)
```

# **4. Winners Using Tuple And Frozenset**

**create winners list**

```python
winners = [("vishnu", "Dance"),
           ("kowshik", "Coding"),
           ("kiruthik", "Singing")]
```

**Convert into frozenset**

```python
winner_set = frozenset(winners)
```

**Display Frozenset**

```python
print("Winners Frozenset:", winner_set)
```
